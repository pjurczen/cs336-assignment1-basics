```python
import regex as re
```

```python
special_token = re.escape("<|endoftext|>")
testtext: str = "some text that i'll pre-tokenize<|endoftext|>some other text i'll pre-tokenize"
delimiter: str = "|".join([special_token])
re.split(delimiter, testtext)
```

```python
def split_data(input: str) -> list[str]:
    return re.split(delimiter, input)
```
